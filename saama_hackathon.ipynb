```
import pandas as pd
import requests
```

```
def getSubjectIds():
  url = 'https://pyhack-dot-pharmanlp-177020.uc.r.appspot.com/api/1/StudyHack/ae/subject/list'
  return requests.get(url).json()["data"]

def getDataForSubject(ids):
  dic1 = []
  for id in ids:
    url = 'https://pyhack-dot-pharmanlp-177020.uc.r.appspot.com/api/1/StudyHack/ae/subject/'+ str(id) + '/list'
    dic1.extend(requests.get(url).json()["data"])
  
  dic2 = []
  for id in ids:
    url = 'https://pyhack-dot-pharmanlp-177020.uc.r.appspot.com/api/1/StudyHack/cm/subject/'+ str(id) + '/list'
    dic2.extend(requests.get(url).json()["data"])
  return dic1, dic2
```

```
def fetchDataFrame(dictionary):
  return pd.DataFrame(dictionary)
```
